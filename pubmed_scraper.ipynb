```python
import bs4
import requests
import csv

url = 'https://pubmed.ncbi.nlm.nih.gov/?term={0}'.format(input('Enter term : '))

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)\
AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}

datas=[]    

for page in range(1,21):

    req = requests.get(url+'&page='+str(page), headers= headers)
    soup = bs4.BeautifulSoup(req.text, 'html.parser')
                                             
    product = soup.findAll('article', 'full-docsum')
            
    for x in product:
    
            pmid = x.find('span','docsum-pmid').text
            author = x.find('span','docsum-authors full-authors').text
            citation = x.find('span','docsum-journal-citation full-journal-citation').text
            
            datas.append([pmid, author, citation])
        
            print('PMID             :',pmid)
            print('Author(s)        :',author)
            print('Citation         :',citation,'\n')
                 
            with open('pubmed.csv', 'w', newline='', encoding='utf-8') as file:
                writer = csv.writer(file)
                headings = ['PMID','Author(s)','Citation']
                writer.writerow(headings)
                for data in datas:
                    writer.writerow(data)

        
                 
```
